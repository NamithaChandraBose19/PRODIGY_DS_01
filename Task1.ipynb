```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
!pip install xlrd
```

```python
# Load the dataset
file_path = 'API_SP.POP.TOTL_DS2_en_excel_v2_3401611.xls'
data = pd.read_excel(file_path, sheet_name='Data', skiprows=3)
```

```python
# Display the table
print("Table preview:")
display(data.head())
```

```python
# Drop columns with all NaN values
data = data.dropna(axis=1, how='all')
```

```python
# Fill NaN values in relevant columns if necessary (e.g., forward fill)
data = data.fillna(method='ffill')
```

```python
# Rename columns for easier access
data.columns = [col.strip() for col in data.columns]
```

```python
# Summary Statistics
print("\nSummary Statistics:")
display(data.describe(include='all')) 
```

```python
# Data Information
print("\nData Information:")
display(data.info())
```

## 1. Top 20 Countries by Population (Most Recent Year)
**Purpose**: To visualize the top 20 countries by population for the most recent year, highlighting global population distribution. <br>
**Use Case**: Useful for comparing population sizes across leading countries to understand demographic trends and resource allocation needs.

```python
# Filter the data to select the most recent year for each country
latest_year = data.columns[-1]  # Assumes the last column is the most recent year

# Extract relevant columns
country_population = data[['Country Name', 'Country Code', latest_year]].copy()

# Rename columns for clarity
country_population.columns = ['Country', 'Country Code', 'Population']

# Sort by population for better visualization
country_population = country_population.sort_values(by='Population', ascending=False)

# Select top N countries for better readability (e.g., top 20)
top_countries = country_population.head(20)

# Set the plotting style
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(14, 8))
bar_plot = sns.barplot(x='Population', y='Country', data=top_countries, palette='viridis')

# Add labels and title
bar_plot.set_title('Top 20 Countries by Population (Most Recent Year)')
bar_plot.set_xlabel('Population')
bar_plot.set_ylabel('Country')

# Show the plot
plt.show()
```

## 2. Bar Plot for Population Growth Over Time (Continuous Variable)
**Purpose**: To visualize the population growth of a specific country over several years. <br>
**Use Case**: Track how the population of a single country has changed over time

```python
# Extract population data for a specific country
country = 'India'  # Example country
country_data = data[data['Country Name'] == country]

# Extract relevant columns (years)
years = list(map(str, range(1960, 2024)))
population_data = country_data[years].melt(var_name='Year', value_name='Population')
population_data['Year'] = pd.to_datetime(population_data['Year'], format='%Y')

# Create the line plot for population growth
plt.figure(figsize=(14, 8))
sns.lineplot(x='Year', y='Population', data=population_data, marker='o', color='blue')
plt.title(f'Population Growth of {country} Over Time')
plt.xlabel('Year')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## 3. Grouped Bar Plot for Population by Region (Categorical and Continuous Variables)
**Purpose**: To compare population sizes across different regions for multiple years. <br>
**Use Case**: Show how populations are distributed across various regions for specific years.

```python
# Extract relevant columns for selected years and countries
years = ['2000', '2020']
data_years = data[['Country Name'] + years].copy()
data_years.columns = ['Country', 'Population_2000', 'Population_2020']

# Filter for a few selected countries (example: India, China, United States, Canada, Brazil)
selected_countries = ['India', 'China', 'United States', 'Canada', 'Brazil']
data_filtered = data_years[data_years['Country'].isin(selected_countries)]

# Melt the DataFrame for a grouped bar plot
data_melted = data_filtered.melt(id_vars=['Country'], var_name='Year', value_name='Population')

# Create a grouped bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Country', y='Population', hue='Year', data=data_melted, palette='viridis')
plt.title('Population by Country for 2000 and 2020')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.legend(title='Year')
plt.show()
```

## 4.Histogram of Population Ranges (Continuous Variable)
**Purpose**: To visualize the distribution of population sizes across countries. <br>
**Use Case**: Show how many countries fall into different population size ranges.

```python
# Extract population data for the most recent year
latest_year = '2023'
population_data = data[['Country Name', latest_year]].copy()
population_data.columns = ['Country', 'Population']

# Create a histogram for population distribution
plt.figure(figsize=(14, 8))
plt.hist(population_data['Population'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Population Sizes Across Countries in 2023')
plt.xlabel('Population Size')
plt.ylabel('Number of Countries')
plt.grid(True)
plt.show()
```
